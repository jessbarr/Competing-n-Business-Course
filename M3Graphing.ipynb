```python
fig,ax = plt.subplots()

x = df['Month-Year']
y1 = df['Private Construction']
y2 = df['Public Construction']

ax.plot(x,y1,color='b',linewidth=.5,label='Private Construction')
ax.plot(x,y2,color='g',linewidth=.5,label='Public Construction')

ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True, direction='in')
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True, direction='in')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel('Month-Year',fontsize='14',fontname='Times New Roman')
ax.set_ylabel('Spending',fontsize='14',fontname='Times New Roman')
ax.set_xticks(x[::8])

ax.axvspan(71,90,ymin=0,ymax = 300,facecolor='k',alpha=0.1)
ax.text('Sep-08',80000,'The Great\nRecession',fontsize=14,verticalalignment='center',horizontalalignment='center',fontname='Times New Roman')
fig.legend(fontsize=12)

fig.set_size_inches(14,6)
fig.suptitle("Construction Value from January 2002 - February 2014 \n(Millions of Dollars)",fontsize='18',fontname='Times New Roman')
fig.savefig('constructionplot.jpg',dpi=600)
```

Private Construction had its highest at 84,649 builds and its lowest of 32,069 builds.  Public Construction's highest amount of builds was 32,069 which was only 283 less than Private Construction's lowest amount.  This minimal difference happened in 2010.  Private Construction had its largest drop in 2010 while Public Construction had its highest numbers. This extreme drop in spending was due to the Great Recession which occurred from December 2007 to June 2009. Private Construction began to slowly recover in 2011 but in 2014 still had not reached its historical peak.  
Both follow the same seasonality and cylicality with the most builds occuring in the warm months and the lowest during the cold months every year.
